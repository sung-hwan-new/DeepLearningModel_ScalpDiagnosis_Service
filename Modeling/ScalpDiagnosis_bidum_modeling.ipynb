# 실행코드

```python
# Focal Loss 클래스 추가
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss
```

```python
# GPU 사용 가능 시 GPU, 아니면 CPU 사용
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
# 하이퍼파라미터 설정: 배치 사이즈와 랜덤 시드 설정
# 재현성을 위한 랜덤시드 고정
hyper_param_batch = 6
random_seed = 100
random.seed(random_seed)
torch.manual_seed(random_seed)
```

```python
# 두피 증상에 대한 클래스 수 설정
num_classes = 3

# 사용할 EfficientNet 모델 이름과 훈련 데이터 폴더 설정
model_name = 'efficientnet-b1'
train_name = 'model_bidum_b1_focal-loss_cosine_lr=1e-3_transform-gray-train-val_3-class'

# 훈련된 모델 가중치를 저장할 경로 설정
PATH = './drive/MyDrive/241122_final_project/data/scalp_weights/'
```

```python
# 훈련, 검증, 테스트 데이터 경로 설정
data_train_path = './data/training'
data_validation_path = './data/validation'
data_test_path = './data/test'
```

```python
# EfficientNet 모델에 맞는 입력 이미지 사이즈 가져오기
image_size = EfficientNet.get_image_size(model_name)
print(f'모델: {model_name}, 이미지 사이즈: {image_size}')
```

```python
# EfficientNet 모델 불러오기 및 출력 클래스 수 설정
model = EfficientNet.from_pretrained(model_name, num_classes=num_classes)
model = model.to(device)
```

```python
# 이미지 크기 조정 함수 추가
def resize_with_aspect_ratio(image, target_size):
    w, h = image.size
    aspect_ratio = w / h
    if aspect_ratio > 1:
        new_w = target_size
        new_h = int(target_size / aspect_ratio)
    else:
        new_h = target_size
        new_w = int(target_size * aspect_ratio)
    return image.resize((new_w, new_h), Image.BICUBIC)
```

```python
# 데이터 증강 및 변환 설정 (훈련 데이터용)
transforms_train = transforms.Compose([
    transforms.Lambda(lambda img: resize_with_aspect_ratio(img, target_size=240)),
    transforms.RandomHorizontalFlip(p=0.5),  # 50% 확률로 좌우 반전
    transforms.RandomVerticalFlip(p=0.5),  # 50% 확률로 상하 반전
    transforms.Lambda(lambda x: x.rotate(90)),  # 90도 회전
    transforms.RandomRotation(10),  # 10도 이내에서 랜덤 회전
    transforms.ColorJitter(brightness=0.1, contrast=0.1),  # 컬러 색상 변경
    transforms.Grayscale(num_output_channels=3),  # 그레이스케일 변환 후 RGB로 채널 복원
    transforms.ToTensor(),  # 텐서로 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 정규화
])

# 데이터 증강 및 변환 설정 (검증 데이터용)
transforms_val = transforms.Compose([
    transforms.Lambda(lambda img: resize_with_aspect_ratio(img, target_size=240)),
    transforms.Grayscale(num_output_channels=3),  # 흑백 변환 후 3채널로 맞춤
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 정규화
])
```

```python
# 훈련 및 검증 데이터셋 불러오기
train_data_set = datasets.ImageFolder(data_train_path, transform=transforms_train)
val_data_set = datasets.ImageFolder(data_validation_path, transform=transforms_val)
```

```python
# DataLoader 설정: 데이터를 배치로 나누고, 학습 및 검증 시 사용
dataloaders, batch_num = {}, {}
dataloaders['train'] = DataLoader(train_data_set, batch_size=hyper_param_batch, shuffle=True, num_workers=4)
dataloaders['val'] = DataLoader(val_data_set, batch_size=hyper_param_batch, shuffle=False, num_workers=4)
```

```python
# 훈련 및 검증 데이터의 배치 수 계산
batch_num['train'], batch_num['val'] = len(train_data_set), len(val_data_set)

print('배치 크기: %d,  훈련/검증: %d / %d' % (hyper_param_batch, batch_num['train'], batch_num['val']))
```

```python
# 클래스 이름 출력 (예: 증상 종류)
class_names = train_data_set.classes
print(class_names)
```

```python
# 오분류 정보 저장 함수 추가
def save_misclassified(misclassified):
    save_path = './drive/MyDrive/241122_final_project/data/error(bidum)/'
    os.makedirs(save_path, exist_ok=True)
    filename = os.path.join(save_path, 'error(bidum)(efficientnet_b1_focal-loss_cosine_lr=1e-3_transform-gray-train-val_3-class).csv')
    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['epoch', 'filename', 'true_label', 'predicted_label']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for item in misclassified:
            writer.writerow(item)
    print(f"All misclassified images information saved to {filename}")
```

```python
# 모델 훈련 함수 정의
def train_model(model, criterion, optimizer, scheduler, num_epochs=25):
    start_time = time.time()  # 시작 시간 기록
    since = time.time()
    best_acc = 0.0  # 최고 정확도 기록을 위한 변수
    best_model_wts = copy.deepcopy(model.state_dict())  # 최상의 모델 가중치 저장

    # 각 에포크별 손실 및 정확도 저장을 위한 리스트
    train_loss, train_acc, val_loss, val_acc = [], [], [], []
    all_misclassified = []

    # 에포크 수만큼 훈련 반복
    for epoch in range(num_epochs):
        print('에포크 {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        epoch_start = time.time()

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # 훈련 모드
            else:
                model.eval()  # 평가 모드

            running_loss = 0.0
            running_corrects = 0
            num_cnt = 0

            for inputs, labels in tqdm(dataloaders[phase]):  # 데이터 로드, tqdm 추가
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()  # 기울기 초기화

                with torch.set_grad_enabled(phase == 'train'):  # 학습 시 기울기 계산
                    outputs = model(inputs)  # 모델에 입력

                    _, preds = torch.max(outputs, 1)  # 예측값
                    loss = criterion(outputs, labels)  # 손실 계산

                    if phase == 'train':
                        loss.backward()  # 기울기 역전파
                        optimizer.step()  # 가중치 갱신

                running_loss += loss.item() * inputs.size(0)  # 손실 합산
                running_corrects += torch.sum(preds == labels.data)  # 정확도 계산
                num_cnt += len(labels)

                # 오분류 정보 수집 (검증 단계에서만)
                if phase == 'val':
                    for i in range(len(labels)):
                        if preds[i] != labels[i]:
                            all_misclassified.append({
                                'epoch': epoch,
                                'filename': val_data_set.samples[num_cnt - len(labels) + i][0],
                                'true_label': class_names[labels[i].item()],
                                'predicted_label': class_names[preds[i].item()]
                            })

            if phase == 'train':
                scheduler.step()  # 학습률 스케줄러 갱신

            # 에포크 손실 및 정확도 계산
            epoch_loss = float(running_loss / num_cnt)
            epoch_acc  = float((running_corrects.double() / num_cnt).cpu() * 100)

            if phase == 'train':
                train_loss.append(epoch_loss)
                train_acc.append(epoch_acc)
                scheduler.step()  # 학습 단계 끝에 스케줄러 호출
            else:
                val_loss.append(epoch_loss)
                val_acc.append(epoch_acc)

            print('{} Loss: {:.4f} Acc: {:.4f}'.format(phase, epoch_loss, epoch_acc))

            # 검증에서 최고 정확도일 경우 모델 저장
            if phase == 'val' and epoch_acc > best_acc:
                best_idx = epoch
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())
                print('==> 최고의 모델 저장 - %d / %.1f' % (best_idx, best_acc))

        # 현재 학습률 출력
        current_lr = optimizer.param_groups[0]['lr']
        print(f'Current Learning Rate: {current_lr:.6f}')

        epoch_end = time.time() - epoch_start
        print('Training epochs {} in {:.0f}m {:.0f}s'.format(epoch, epoch_end // 60, epoch_end % 60))
        print()

    # 훈련 완료 시간 출력
    time_elapsed = time.time() - start_time
    print('훈련 완료 시간: {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
    print('Best valid Acc: %d - %.1f' %(best_idx, best_acc))

    # 최고 성능 모델의 가중치 불러오기
    model.load_state_dict(best_model_wts)

    # 모델 및 가중치 저장
    torch.save(model, PATH + 'aram_'+train_name+'.pt')
    torch.save(model.state_dict(), PATH + 'president_aram_'+train_name+'.pt')
    print('모델 저장 완료')

    end_sec = time.time() - start_time
    end_times = str(datetime.timedelta(seconds=end_sec)).split('.')
    end_time = end_times[0]
    print("end time :", end_time)

    # 오분류 정보 저장
    save_misclassified(all_misclassified)

    return model, best_idx, best_acc, train_loss, train_acc, val_loss, val_acc
```

```python
# 손실 함수 및 옵티마이저 설정
# criterion = nn.CrossEntropyLoss()  # 기존 설정 주석 처리

# Focal Loss 사용
criterion = FocalLoss(gamma=2)
optimizer_ft = optim.Adam(model.parameters(), lr=1e-3)

# CosineAnnealingLR 스케줄러 설정
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer_ft, T_max=10, eta_min=1e-5)
```

```python
# 에포크 수 설정하고 훈련 시작
num_epochs = 5
train_model(model, criterion, optimizer_ft,scheduler, num_epochs=num_epochs)
```

# 하이퍼파라미터 값 찾기

```python
# 그리드 탐색 (최적의 하이퍼파라미터 값 찾기)

import torch
import torch.optim as optim
from efficientnet_pytorch import EfficientNet

# 탐색할 하이퍼파라미터 목록
learning_rates = [1e-4, 1e-3, 1e-2]
t_max_values = [5, 10, 20]
eta_min_values = [1e-6, 1e-5]

best_acc = 0.0
best_hyperparams = {}

for lr in learning_rates:
    for t_max in t_max_values:
        for eta_min in eta_min_values:
            print(f"Training with lr={lr}, T_max={t_max}, eta_min={eta_min}")

            # 모델 초기화
            model = EfficientNet.from_pretrained('efficientnet-b1', num_classes=3)
            model = model.to(device)

            # 옵티마이저 및 스케줄러 설정
            optimizer = optim.Adam(model.parameters(), lr=lr)
            scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=t_max, eta_min=eta_min)

            # 훈련 시작 (train_model 함수로 대체)
            model, _, epoch_acc, _, _, _, _ = train_model(model, criterion, optimizer, scheduler, num_epochs=5)

            # 최적 성능 저장
            if epoch_acc > best_acc:
                best_acc = epoch_acc
                best_hyperparams = {'lr': lr, 'T_max': t_max, 'eta_min': eta_min}

print(f"Best accuracy: {best_acc}")
print(f"Best hyperparameters: {best_hyperparams}")
```

# 모델의 정답/오답 클래스 확률 확인

```python
import torch
import os
import csv
from torchvision import transforms
from PIL import Image
```

```python
# GPU 사용 여부 확인
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# 모델 불러오기 (저장된 모델 경로 사용)
model = torch.load("./drive/MyDrive/241122_final_project/data/scalp_weights/aram_model_bidum_b1_focal-loss_cosine_lr=1e-3_transform-gray-train-val_3-class.pt")  # 모델 경로 수정
model = model.to(device)
model.eval()  # 평가 모드로 전환
```

```python
# 이미지 전처리 설정
transform = transforms.Compose([
    transforms.Resize((240, 240)),  # 모델 입력 크기에 맞게 조정
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 이미지 정규화
])
```

```python
# 이미지가 저장된 폴더 경로 설정
image_folder = './data/validation/[원천]비듬_2.중증'  # 이미지 폴더 경로 수정
```

```python
# 결과를 저장할 CSV 파일 경로
csv_file = './drive/MyDrive/241122_final_project/data/probabilities/241023_bidum_top_two_class_probabilities-class-2.csv'
```

```python
# 클래스 이름 (모델이 학습한 클래스 이름을 사용)
class_names = ['[원천]비듬_0.양호', '[원천]비듬_1.경증_중등도', '[원천]비듬_2.중증']  # 클래스 이름 수정
```

```python
# CSV 파일을 만들기 위한 설정
header = ['image', 'top_class', 'top_probability', 'second_class', 'second_probability']  # 헤더 설정

with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # 헤더 작성

    # 이미지 폴더의 모든 파일에 대해 예측
    for image_name in os.listdir(image_folder):
        if image_name.endswith(('.png', '.jpg', '.jpeg')):  # 이미지 파일 확장자 필터링
            image_path = os.path.join(image_folder, image_name)

            # 이미지 불러오기
            image = Image.open(image_path).convert('RGB')  # 이미지를 RGB로 변환
            image = transform(image).unsqueeze(0).to(device)  # 이미지 전처리 및 배치 추가

            # 모델에 입력하여 클래스별 확률 계산
            with torch.no_grad():  # 평가 모드에서 기울기 계산 방지
                outputs = model(image)
                probabilities = torch.nn.functional.softmax(outputs, dim=1)  # softmax로 확률 변환

            # 각 이미지에 대한 클래스별 확률값 저장
            prob_values = probabilities.cpu().numpy()[0]  # 확률값을 numpy 배열로 변환

            # 상위 두 개의 확률 및 해당 클래스 찾기
            top_two_indices = prob_values.argsort()[-2:][::-1]  # 상위 2개의 인덱스 (내림차순)
            top_class = class_names[top_two_indices[0]]  # 첫 번째 클래스
            top_probability = prob_values[top_two_indices[0]]  # 첫 번째 확률
            second_class = class_names[top_two_indices[1]]  # 두 번째 클래스
            second_probability = prob_values[top_two_indices[1]]  # 두 번째 확률

            # 이미지 이름과 상위 두 클래스, 확률을 CSV로 저장
            row = [image_name, top_class, top_probability, second_class, second_probability]
            writer.writerow(row)
```
